# Google Maps API

```python
from blueflow import notebooks

import math
import matplotlib.pyplot as plt
import numpy as np

from blue_options import string
from blue_objects import objects

from roofai.host import signature
from roofai.google_maps.api.static import get as get_static_image
from roofai.logger import logger

logger.info(f"{'.'.join(signature())},\nbuilt on {string.pretty_date()}")
```

```python
object_name = objects.unique_object("fascinating-magic")
object_path = objects.object_path(object_name)

filename = objects.path_of(
    filename=f"{string.timestamp()}.png",
    object_name=object_name,
    )
```

---

```python
lat = 53.343318
lon = -2.650661
zoom = 20
maptype = "satellite"
size = "647x640"
```

```python
success, image, metadata = get_static_image(
        lat=lat,
        lon=lon,
        zoom=zoom,
        maptype=maptype,
        size=size,
        filename=filename,
        verbose=True,
    )
assert success
```

```python
notebooks.imshow(filename)
```
